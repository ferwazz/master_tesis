```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
data=pd.read_csv('/home/fernando/Desktop/Fernando/TAOSII/xo-3-r-2016-10-21.csv',delimiter=" ")
```

```python
data
```

```python
plt.figure(figsize=(15,10))
plt.plot(np.cumsum(data['Texp'])/60,data['Mag_b']-data['Mag_a'],'o',markersize=4,linewidth=1,label='R-2016-10-21')
plt.xlabel('Tiempo (min)',size=25)
plt.ylabel('Diferencia en magnitudes',size=25)
plt.xticks(size=20)
plt.yticks(size=20)
plt.legend(fontsize=20)
plt.savefig('../reports/figures/transitoXO3.png',dpi=300)
```
